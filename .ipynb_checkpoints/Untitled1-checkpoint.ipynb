```python
## 원하는 장치 입력하면 장치와 센서 맥주소 리턴
import plotly.graph_objs as go
def dev_sen_mac(search,api_key,api_secret,limit_up,limit_down):
    url = 'https://oa.tapaculo365.com/tp365/v1/channel/get_lst'
    param ={
    'search' : search,
    'api_key' : api_key,         #4528458
    'api_secret' : api_secret    #'3fa6d40e4461cd250a683b86eed42bad'
    }
    req = requests.get(url,params=param)
    device_mac = req.json()['rows'][0]['device_mac']
    sensor_mac = req.json()['rows'][0]['sensor_mac']
    def device_24info():
        url2 = 'https://oa.tapaculo365.com/tp365/v1/channel/get_recentdata'
        param2 ={
        'device_mac' : device_mac,
        'sensor_mac' : sensor_mac,
        'api_key' : api_key,
        'api_secret' : api_secret
        }
        req2 = requests.get(url2,params=param2,headers={'User-Agent': 'Mozilla/5.0'})
        data = req2.json()
        df = pd.DataFrame(data['rows'])
        df.timestamp = list(map(lambda x: datetime.datetime.fromtimestamp(int(x)).strftime('%Y-%m-%d %H:%M:%S'),df.timestamp ))
        df = df.filter(['timestamp','ch1','ch2'])
        df.timestamp = list(map(lambda x: pd.to_datetime(x,format='%Y-%m-%d %H:%M:%S'),df.timestamp))
        df.ch1 = df.ch1.astype(float)
        df.ch2 = df.ch2.astype(float)
        df.columns = ['Time','Temp','Humidity']
        exceed_count = len(df.Temp.where(df.Temp > limit_up).dropna()) + len(df.Temp.where(df.Temp < limit_down).dropna())
        exceed_percent = exceed_count / len(df.Temp) * 100
        
        # 임계치 초과한 부분 나누기 (인덱스 번호로 나누었음)
        start = []  # 임계치가 초과한 시점 담을 리스트
        end = [] # 임계치가 다시 초과하지 않게 되는 시점을 담을 리스트
        d = df[(df.Temp > limit_up) | (df.Temp < limit_down)] # 임계치를 초과한 측정기록
        start.append(d.iloc[0].name) # 임계치 초과 하는 첫 시작점 = 위에서 구한 d의 첫번째 행
        for i in range(len(d)-1):
            if (d.Time.iloc[i+1] -  d.Time.iloc[i]).seconds != 300:
                start.append(d.iloc[i+1].name)
                end.append(d.iloc[i].name)

            else:
                pass
        end.append(d.iloc[-1].name)
        
        # 임계치 초과 한 시간 구하기
        b=0
        c =0
        for i in range(len(start)):
            if end[i] - start[i] == 0:
                b += 300
            else:
                c += (end[i] - start[i] + 1) * 300
        (b+c) /60 / 60 #임계치를 벗어나 있던 시간
        
        df['class'] = 0
        df['Temp_gap'] = 0
        df['Time(sec)'] = 0
        df['gradient'] = 0
        df['class'].iloc[:-1] = list(map(lambda x,y:'cold' if x > y else 'warm', df.Temp.iloc[:-1],df.Temp.iloc[1:]))
        df['Temp_gap'].iloc[:-1]= list(map(lambda x,y:y-x if x > y else y-x, df.Temp.iloc[:-1],df.Temp.iloc[1:]))
        df['Time(sec)'].iloc[:-1]= list(map(lambda x,y:(y-x).seconds if x > y else (y-x).seconds, df.Time.iloc[:-1],df.Time.iloc[1:]))
        df['gradient'].iloc[:-1]= list(map(lambda x,y:x/y if x > y else x/y, df.Temp_gap.iloc[:-1],df['Time(sec)'].iloc[:-1]))
        df = df.iloc[:-1].copy()
        
        df3 = df.groupby('class').mean().gradient * 60

        max_min = []
        max_min.append(df.Temp.max())
        max_min.append(df.Temp.min())
        df3.loc['Maxtemp'] = df.Temp.max()
        df3.loc['Mintemp'] = df.Temp.min()
        df3.loc['Meantemp'] = df.Temp.mean()


        df3
        df4 = pd.DataFrame(df3)
        df4.columns = [search]
        df4.rename(index={'cold':'cold_temp_permin','warm':'warm_temp_permin'},inplace=True)
        df4.loc['exceed_time'] = (b+c) /60 / 60
        df4.loc['exceed_count'] = exceed_count
        df4.loc['exceed(%)'] = exceed_percent
        
        up_limit = df[(df.Temp > limit_up) | (df.Temp < limit_down)]
        in_limit = df[(df.Temp <= limit_up) & (df.Temp >= limit_down)]

        fig = go.Figure()
        fig.add_trace(go.Scatter(x=df.Time,y=df.Temp,mode='lines',name='그래프',line=dict(color='black')))
        fig.add_trace(go.Scatter(x=up_limit.Time,y=up_limit.Temp,mode='markers',name='임계치 초과'))
        fig.add_trace(go.Scatter(x=in_limit.Time,y=in_limit.Temp,mode='markers',marker=dict(color='Blue'),name='임계치 이내'))
        fig.add_trace(go.Scatter(x=df.Time,y=[limit_up] * len(df.Temp),mode='lines',line=dict(color='Orange'),name='임계치'))
        fig.add_trace(go.Scatter(x=df.Time,y=[limit_down] * len(df.Temp),mode='lines',line=dict(color='Orange'),name='임계치'))
        fig.update_layout(width=1000,height=600,title=dict(text = search,font=dict(size= 30)),yaxis_title='온도',font=dict(size=15))
        fig.show()


        
        #print(df)
        return df4
    
    return device_24info()
```
